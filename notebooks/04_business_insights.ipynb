## Business Insights & Recommendations

### Insight 1: High churn among customers with short-term contracts
Customers on shorter contract lengths show significantly higher churn rates.

**Recommendation:**
- Offer discounts or benefits for upgrading to longer-term contracts.
- Introduce loyalty incentives for customers nearing contract renewal.

---

### Insight 2: New customers are at higher risk of churn
Customers with lower tenure are more likely to churn compared to long-term customers.

- Improve onboarding experience during the first year.
- Provide proactive customer support for new customers.
